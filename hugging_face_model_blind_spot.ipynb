```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "Nanbeige/Nanbeige4-3B-Base"

tokenizer = AutoTokenizer.from_pretrained(
    model_name,
    use_fast=False,
    trust_remote_code=True
)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto",
    trust_remote_code=True
)

prompt = "I have 8 eggs, I broke 2 eggs, fried 2 eggs and I ate 2 eggs. How many eggs do I have left?"

input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(model.device)

output_ids = model.generate(
    input_ids,
    max_new_tokens=200,
    temperature=0.7,
    do_sample=True
)

resp = tokenizer.decode(output_ids[0][len(input_ids[0]):], skip_special_tokens=True)
print(resp)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "Nanbeige/Nanbeige4-3B-Base"

tokenizer = AutoTokenizer.from_pretrained(
    model_name,
    use_fast=False,
    trust_remote_code=True
)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto",
    trust_remote_code=True
)

prompt = """A farmer bought a wolf, a goat, and a cabbage.

He must cross a river using a boat that can carry only the farmer and one item at a time.

If left alone together without the farmer:
- the wolf will eat the goat
- the goat will eat the cabbage

How can the farmer transport all three safely across the river?"""

input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(model.device)

output_ids = model.generate(
    input_ids,
    max_new_tokens=1000,
    temperature=0.7,
    do_sample=True
)

resp = tokenizer.decode(output_ids[0][len(input_ids[0]):], skip_special_tokens=True)
print(resp)
```
